## Open image and stuff
Here I'll open, resize, convert and play with an image.

First, open and extract some basic information about the image

```python
bsb = Image.open('bsb.jpg') #open image
print("Here some usefull informations about the image:")
print("Format: " + str(bsb.format))
print("Size: "+ str(bsb.size))
print("Mode: " + str(bsb.mode)) 

#Print thumbnail
bsb_thumbnail = bsb.copy()
bsb_thumbnail.thumbnail((256,256))
bsb_thumbnail
```

Now I'll convert in grayscale, rotate and crop the image.

Then copy and paste a region of this gray area in other region in the first image

```python
bsb_gray = bsb.copy()
bsb_gray = bsb_gray.convert("L")
#bsb_gray = bsb_gray.thumbnail((128,256)) #creat a thumbnail 
bsb_gray = bsb_gray.rotate(180).crop((450,0,900,420))  #crop(left,upper,right,lower)

bsb_merged = bsb.copy()
bsb_merged.paste(bsb_gray, box=(450,0,900,420)) #Past one image in other
bsb_merged
```
